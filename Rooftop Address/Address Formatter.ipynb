```python
import csv
```

```python
with open('208.csv') as in_file:
    rows_read = in_file.readlines()
    with open('208_addr_seperated.csv', 'w') as out_file:
        for row in rows_read:
            print(row[:-1])
            nr_index = row.find('號') + 1
            out_file.write(row[:nr_index] + ',' + row[nr_index:])
```

```python
with open('addr.csv','w') as out_file:
    out_file.write('idx,sec,addr\n')
    
    with open('226.csv',encoding='utf') as in_file:
        rows = csv.reader(in_file)
        for index, row in enumerate(rows):
            if index == 0:
                continue
            nr_index = row[5].find('號') + 1
            out_file.write(str(index) + ',' + row[4] + ',' + row[5][:nr_index] + '\n')
        
    with open('208.csv') as in_file:
        rows = csv.reader(in_file)
        for index, row in enumerate(rows):
            if index == 0:
                continue
            nr_index = row[2].find('號') + 1
            out_file.write(str(index+226) + ',' + row[1] + ',' + row[2][:nr_index] + '\n')
```
