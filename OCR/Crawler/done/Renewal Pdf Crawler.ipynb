```python
import requests
from bs4 import BeautifulSoup as BS
import re
from tqdm import tqdm_notebook
import urllib3
import ssl
```

```python
#ssl._create_default_https_context = ssl._create_unverified_context
#url = 'https://uro.gov.taipei/News_Content.aspx?n=AF4D78CF694B745C&sms=CCC54E6046E281ED&s=0C0C4F367526A01E'
url = 'https://uro.gov.taipei/News_Content.aspx?n=AF4D78CF694B745C&sms=CCC54E6046E281ED&s=B014264347C741B4'
response = requests.get(url, verify=False)
soup = BS(response.text,'html5lib')
```

```python
soup.find_all('a',text=re.compile(r'書'))
```

```python
files_done = 16
links=soup.find_all('a',text=re.compile(r'書'))[2+files_done:-1]
```

```python
file_links = [link['href'] for link in links][:]
```

```python
tqdm_notebook.monitor_interval=0
for file_link in tqdm_notebook(file_links):
    filename = file_link.split('/')[-1]
    response = requests.get(file_link)
    with open(filename,'wb') as outf:
        outf.write(response.content)
```

```python
tqdm.monitor_interval=0
with open('urls.txt','w') as outf:
    outf.write('\n'.join(file_links))
```
